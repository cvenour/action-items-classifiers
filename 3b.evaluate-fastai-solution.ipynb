```python
def evaluateResults(classifierLabels, humanLabels):
    
    numPredictions = len(classifierLabels)
    truePositives = 0
    trueNegatives = 0

    falsePositives = 0
    falseNegatives = 0
    numCorrect = 0
    numWrong = 0

    for i in range(len(humanLabels)): # assuming the lists are of the same length
        if (humanLabels[i]==1):
            if (classifierLabels[i] == 1):
                truePositives = truePositives + 1
                numCorrect = numCorrect + 1
            elif(classifierLabels[i] == 0):
                falseNegatives = falseNegatives + 1
                numWrong = numWrong + 1
        elif(humanLabels[i]==0):
            if (classifierLabels[i] == 0):
                trueNegatives = trueNegatives + 1
                numCorrect = numCorrect + 1
            elif(classifierLabels[i]==1):
                falsePositives = falsePositives + 1
                numWrong = numWrong + 1


    print("true positives:", truePositives)
    print("false negatives:", falseNegatives)
    print("false positives:", falsePositives)
    print()

    accuracy = numCorrect/numPredictions
    precision = truePositives/(truePositives + falsePositives)
    recall = truePositives/(truePositives + falseNegatives)
    
    return (accuracy,precision,recall)
```

```python
df_test = pd.read_csv('./action-items-data/data-originals/test_balanced_labelled_4.csv')

testUtterances = df_test['utterance'].tolist()
testLabels = df_test['label'].tolist()
num_test_utterances = len(testUtterances)
print(num_test_utterances)
```

```python
print(testUtterances[0:10])
```

```python
Ypredictions = []
for utt in testUtterances:
    answer = learn2.predict(utt)
    someprediction = np.asarray(answer[2])
    somelabel = np.argmax(someprediction)
    Ypredictions.append(somelabel)
    
```

```python
print(len(Ypredictions))
print(Ypredictions)
```

```python
print(len(testLabels))
print(testLabels)
```

```python
acc,prec,recall = evaluateResults(Ypredictions, testLabels)
print("accuracy:", acc)
print("precision:", prec)
print("recall:", recall)
```
